```python
import pandas as pd

# Load the uploaded Excel file
file_path = 'Patients.xlsx'
df = pd.read_excel(file_path)

# Print the header (first few rows) of the data
header = df.head()

print(header)
```

```python
# Mapping dictionary from const_provider table
provider_mapping = {
    "N/A": 0,
    "Varun": 1,
    "Menga": 2,
    "Rogerson": 3,
    "Voloshin": 4,
    "Puvanesarajah": 5,
    "Molinari": 6,
    "Mannava": 7,
    "Rubery": 8,
    "Daley-Lindo": 9,
    "Aktas": 10,
    "Lander": 11
}

# Replace the Provider column values using the mapping dictionary
df['Provider'] = df['Provider'].map(provider_mapping)

# Display the modified dataframe
df.head()
```

```python
distinct_providers = df['Provider'].unique()
distinct_providers
```

```python
# Mapping dictionary for Research/Patient values
research_patient_mapping = {
    "Medtronic": 1,
    "P30": 2,
    "Validation": 3,
    "Control": 4,
    "Patient": 5,
    "Champp": 6
}

# Replace the Research/Patient column values using the mapping dictionary
df['Research/Patient'] = df['Research/Patient'].map(research_patient_mapping)

# Display the modified dataframe
df.head()
```

```python
# Mapping dictionary for Body Part values
body_part_mapping = {
    "Spine": 1,
    "Spine & Shoulder": 2,
    "Spine (Lumbar)": 3,
    "Shoulder": 4,
    "nan": 0,
    "Spina Bifida": 6,
    "VA-patient": 7
}

# Replace the Body Part column values using the mapping dictionary
df['Body Part'] = df['Body Part'].map(body_part_mapping)

# Display the modified dataframe
df.head()
```

```python
# Mapping dictionary for Involved Side values
involved_side_mapping = {
    "nan": 0,
    "Right": 1,
    "Left": 2
}

# Replace the Involved Side column values using the mapping dictionary
df['Involved Side'] = df['Involved Side'].map(involved_side_mapping)

# Display the modified dataframe
df.head()
```

```python
distinct_providers = df['Dominant Side'].unique()
distinct_providers
```

```python
# Create a function to map values to Dominant Hand and Dominant Leg
def map_dominant_side(value):
    if pd.isna(value) or value in ['missing', 'not on sheet']:
        return (0, 0)
    elif value in ['Right-Hand,Left-Leg', 'Right-Hand, Left-Leg']:
        return (1, 2)
    elif value in ['Left-Hand,Right-Leg', 'Left-Hand, Right-Leg']:
        return (2, 1)
    elif value in ['Right-Hand,Right-Leg', 'Right-Hand, Right-Leg']:
        return (1, 1)
    elif value in ['Left-Hand,Left-Leg', 'Left-Hand, Left-Leg']:
        return (2, 2)
    elif value == 'Left':
        return (2, 2)
    elif value == 'Right':
        return (1, 1)
    else:
        return (0, 0)  # Default case

# Apply the function to the Dominant Side column and create new columns
df[['Dominant Hand', 'Dominant Leg']] = df['Dominant Side'].apply(map_dominant_side).apply(pd.Series)

# Display the modified dataframe
df.head()
```

```python
# Mapping dictionary for EMG? values
emg_mapping = {
    "No": 0,
    "Yes": 1
}

# Replace the EMG? column values using the mapping dictionary and ensure null values remain as null
df['EMG?'] = df['EMG?'].map(emg_mapping).fillna(pd.NA)

# Display the modified dataframe
df.head()
```

```python
# Mapping dictionary for CDP? values
cdp_mapping = {
    "No": 0,
    "Yes": 1
}

# Replace the CDP? column values using the mapping dictionary and ensure null values remain as null
df['CDP?'] = df['CDP?'].map(cdp_mapping).fillna(pd.NA)

# Display the modified dataframe
df.head()
```

```python
# Rename the columns
df = df.rename(columns={
    'Lab ID': 'lab_id',
    'Age': 'age',
    'Gender': 'gender',
    'Research/Patient': 'type_id',
    'Enrolled Date': 'enrolled_date',
    'Surgical Date': 'surgical_date',
    'Provider': 'provider_id',
    'Body Part': 'body_part_id',
    'Involved Side': 'involved_side',
    'EMG?': 'is_emg',
    'CDP?': 'is_cdp',
    'Dominant Hand': 'dominant_hand',
    'Dominant Leg': 'dominant_leg'
})

# Drop the Dominant Side column
df = df.drop(columns=['Dominant Side'])

# Display the modified dataframe
df.head()
```

```python
# Rename the columns
df = df.rename(columns={
    'Weight': 'weight',
    'Height': 'height',
})

# Display the modified dataframe
df.head()
```

```python
# Create an auto-incrementing id column
df['id'] = range(1, len(df) + 1)

# Move the id column to the front
df = df[['id'] + [col for col in df.columns if col != 'id']]

# Display the modified dataframe
df.head()
```

```python
csv_file_path = 'Patients_with_ID.csv'
df.to_csv(csv_file_path, index=False)
```
